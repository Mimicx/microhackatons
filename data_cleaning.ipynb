```python
#text = 'user_id;star_wars;lord_of_the_rings;harry_potter'
df['index'] = df['user_id;star_wars;lord_of_the_rings;harry_potter']
#df.drop([0, 1])
columns = df.columns[1].split(';')
d2 = []
i = 0
rows = []
for val in range(len(df)): 
    rows.append(df['index'][i].split(';'))                            
    i += 1
    
for rval in range(len(rows)):
    col_num = 0
    data = {}
    for column in columns:
        print(column +' / '+ rows[rval][col_num])
        data[column] = rows[rval][col_num]       
        print(data)
        col_num += 1
    d2.append(data)
        

    


```
